```python
import os
import numpy as np
# from scipy import misc
import matplotlib.pyplot as plt
import pandas as pd
import cv2
# from scipy.signal import find_peaks_cwt
import peakutils #conda install -c conda-forge peakutils
from scipy.signal import savgol_filter
```

```python
# carrega imagem
frame = cv2.imread('img/swell_01.jpg')

img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

c = img[:,0]

c_filt = savgol_filter(c,13,3)
# metodo 2

# thres = 0.02 / np.max(c)
thres = 0.7 # de 0 a 1
min_dist = 12

indexes = peakutils.indexes(c_filt.astype(float)*-1, thres, min_dist)

plt.figure(figsize=(12,8))

plt.imshow(img)
plt.plot(c_filt, range(len(c_filt)))
plt.plot(c_filt[indexes], np.arange(0,len(c_filt))[indexes],'or')
```
